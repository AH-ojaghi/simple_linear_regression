```python
#create dataset with numpy
import numpy as np
X = np.array([1, 2, 3, 4, 6, 9, 10, 15, 13])
Y = np.array([2, 6, 8, 11, 15, 21, 30, 40, 34])

```

```python
#create a function for plotting data using maplotlib library
import matplotlib.pyplot as plt
def plotData(x, y):
    plt.plot(x, y, 'ro', ms=10, mec='k')
    plt.ylabel('eggs')
    plt.xlabel('chicken')
```

```python
#create a function for plotting line
def abline(slope, intercept, epoch):
    """Plot a line from slope and intercept"""
    axes = plt.gca()
    x_vals = np.array(axes.set_xlim(-20,20))
    y_vals = intercept + slope * x_vals
    plt.ylim((-50, 50))
    plt.figure(epoch)
    plt.plot(x_vals, y_vals, '--')
    plt.text(0, -4, 'epoch ' + str(epoch + 1))
```

```python
#visualize data
plotData(X, Y)
```

h(x) = theta0 + theta1 * x

Gradient Descent Formula:

<p align="center">
    <img width="60%" src="https://drive.google.com/uc?id=1AEa5mHzaonQGio3Ay20U4S_4eoA_E2EY">
</p>

```python
#random initialization theta0 and theta1
theta0=1
theta1=0

#hyper parameters and training
m = 9
lr = 0.025
epochs=20
for epoch in range(epochs):
  sum0 = 0
  sum1 = 0
  for i in range(m):
    sum0 = sum0 + (theta0 + theta1*X[i] - Y[i])
    sum1 = sum1 + (theta0 + theta1*X[i] - Y[i]) * X[i]

  theta0 = theta0 - (lr/m)*(sum0)
  theta1 = theta1 - (lr/m)*(sum1)

  plotData(X, Y)
  abline(theta1, theta0, epoch)
```

predict output for x=8

using h(x) = theta0 + theta1 * x

```python
#predict output for x=8
h = theta0 + theta1 * 8
print(h)
```
